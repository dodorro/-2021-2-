그러나 제휴(Affiliation_Spend)의 경우 히트맵상으로는 상관관계가 있어보이지만, pair plot에서 모호한 관계를 보이므로 더 면밀한 분석이 필요하다.

```python
#매출 예상을 위한 선형회귀분석을 실시한다.
Labels = df_media['Sales']
#Y값에 해당하는 Sales를 Labels로 지정한다.
Features = df_media[['Google_Spend', 'Youtube_Spend', 'SNS_Spend', 'Affiliate_Spend']]
#X값에 해당하는 변수들을 Features로 지정한다.
```

```python
print(Labels.shape)
Labels.head()
```

```python
Features.head()
```

```python
# sklearn의 선형회귀분석을 실행하기위해 라이브러리에서 관련 함수를 불러온다.
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(Features, Labels)
print(model.intercept_, model.coef_)
```

```python
#파이썬의 Sklearn 패키지의 선형회귀분석모형은 기본적으로 회귀계수는 제공하지만 회귀계수의 표준오차 및 p-value는 제공하지 않는다.
#따라서 모델의 신뢰도를 확인해보기 위해 statsmodel 패키지를 사용해 회귀계수표를 확인해보겠다.

from statsmodels.formula.api import ols
model_sm = ols(formula = 'Sales ~ Google_Spend + Youtube_Spend + SNS_Spend + Affiliate_Spend', data=df_media).fit()
model_sm.summary()
```

R-square이 0.77수준이므로 모델의 설명력이 높다고 볼 수 있다.

P-value의 경우 모든 변수들이 0.05보다 낮게 나왔으므로 모든 미디어가 매출에 미치는 영향이 있다는 것으로 볼 수 있다.

coef의 경우 상관관계를 나타내는데, 제휴마케팅이 음의 상관관계를 갖는다.

따라서 Affiliate_Spend를 제외한 모델 또한 분석하여 어떤 모델이 최적의 모델인지 검증해봐야 한다.

```python
Labels1 = df_media['Sales']
Features1 = df_media[['Google_Spend', 'Youtube_Spend', 'SNS_Spend']]

model1 = LinearRegression().fit(Features1, Labels1)
print(model1.intercept_, model1.coef_)

model1_sm = ols(formula = 'Sales ~ Google_Spend + Youtube_Spend + SNS_Spend', data=df_media).fit()
model1_sm.summary()
```

Affiliate_Spend 변수가 빠졌음에도 AIC와 BIC의 변화가 미비하다.

AIC와 BIC는 로그우도를 독립변수의 수로 보정한 값으로써 여러 통계 모델들의 성능을 비교할 수 있게 한다.\
값이 작을수록 모델이 우수함을 나타낸다.

Affiliate_Spend 변수가 빠졌음에도 위 수치의 변화가 미비함으로, Affiliate_Spend가 Sales에 미치는 영향이 적다는 것을 알 수 있다. 

위 모델은 최적의 Sales를 구하기 위한 모델이므로 coef값이 음수가 나오는 Affiliate_Spend를 제외해야 한다.

다만, Affiliate_Spend를 제외하기 전에 이를 다시 한 번 확인해볼 필요가 있다.

```python
#displot으로 Affiliate의 그래프를 확인해본다.
sns.distplot(df_media['Affiliate_Spend'])
```

AIC와 BIC는 더 낮아졌지만, log_Affiliate_Spend는 여전히 음의 회기계수를 갖고 있으므로, Affiliate_Spend는 이번 광고 집행 미디어에서는 제외하도록 한다.

```python
#반복문을 통해 모델이 예측할 수 있는 매출의 최대 미디어 믹스 비용을 예상한다.
maxmedia=[] #최대값을 넣을 수 있는 빈 리스트 작성

max= 0 #최대값을 비교하기 위한 변수 선언

for i in range(1, 301):
    Google=498*i #가중치 입력
    for j in range(i,300-i):
        Youtube = 1438*j
        for k in range(j, 301-j):
            SNS = 1678*k
            if i+j+k <=250: #예상비용 이하
                if Google+Youtube+SNS >= max:
                    max = Google+Youtube+SNS
                    maxmedia.append([i,j,k])

print(maxmedia)
```

반복문을 사용하다보니, 가중치가 제일 높은 SNS비용만 사용하라고 컴퓨터는 예측한다. 따라서, 추가적인 미디어 인사이트를 고려해 미디어 믹스를 계획해야한다.

```python
#분석을 위해 열을 재정의한다.
df_target['ConsumePower']= df_target['MntWines'] + df_target['MntFruits'] + df_target['MntMeatProducts']+ df_target['MntFishProducts']+ df_target['MntSweetProducts']
#ComsumePower라는 새로운 열은 소비자의 구매력을 나타내는 지수이다.
df_target['Age']= 2014-df_target['Year_Birth']
#Age는 나이를 보기 원활하게 정수형태로 변환한 값이다.
df_target.rename(columns = {'NumCatalogPurchases':'SubPurchases', 'NumWebPurchases':'OnlinePurchases', 'NumStorePurchases': 'WorkoutTimePerWeek'}, inplace=True)
#각각의 변수명을 변경한다. 특히 NumStorePurchases는 WorkoutTimePerWeek로서, 일주일에 운동하는 시간을 뜻한다.(단위: 시간)
df_target
```

```python
#필요한 열들만 재정의해서 각각의 상관관계를 알아본다.
df_targetTrue=df_target[['Income', 'Age', 'ConsumePower', 'SubPurchases', 'OnlinePurchases', 'WorkoutTimePerWeek']]
corr = df_targetTrue.corr()
sns.heatmap(corr, annot=True, cmap='Reds')
```

인사이트: 유료구독서비스다보니 구매력이 중요할 것이다. 따라서 구매력과 상관관계가 높은 지표들로 클러스터링을 진행한다. Income, SubPurchases, WorkoutTimePerWeek이 후보군이다. 이 중에서 WorkoutTimePerWeek이 랭킹닭컴의 핵심제품인 닭가슴살을 통한 단백질 보충과 가장 직접적인 연관성이 높은 지표이다. 따라서 이를 통해 클러스터링을 진행한다.

```python
sns.pairplot(df_targetTrue[['Income', 'ConsumePower', 'SubPurchases', 'WorkoutTimePerWeek']])
```

```python
from sklearn.cluster import KMeans
#변수에 각각의 값을 입력한다.
X1 = df_targetTrue[['ConsumePower', 'WorkoutTimePerWeek']].values
#빈 리스트 작성
inertia = []
#반복문을 이용해 X1의 k-means++알고리즘을 통해 위의 리스트에 저장한다.
for n in range(1, 11):
    algorithm = (KMeans(n_clusters = n, init='k-means++', n_init = 10, max_iter=300, tol=0.0001, random_state=111, algorithm='elkan'))
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```python
inertia
#값이 들어갔는지 확인한다.
```

```python
#시각화를 통해 적절한 클러스터 개수를 확인한다.
plt.figure(1, figsize=(16,5))
plt.plot(np.arange(1, 11), inertia, 'o')
plt.plot(np.arange(1, 11), inertia, '-', alpha=0.8)
plt.xlabel('Number of Clusters'), plt.ylabel('inertia')
plt.show()
```

```python
#클러스터 개수를 4개로 지정하여 군집화를 진행하고 시각화한다.
algorithm = (KMeans(n_clusters = 4, init='k-means++', n_init = 10, max_iter=300, tol=0.0001, random_state=111, algorithm='elkan'))
algorithm.fit(X1)
labels1 = algorithm.labels_
centroids = algorithm.cluster_centers_

h=0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max()+1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z2 = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])

plt.figure(1, figsize = (15, 7))
plt.clf()
Z2=Z2.reshape(xx.shape)
plt.imshow(Z2, interpolation='nearest', 
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect='auto', origin='lower')

plt.scatter(x='ConsumePower', y='WorkoutTimePerWeek', data=df_targetTrue, c=labels1, s =200)
plt.scatter(x=centroids[:, 0], y=centroids[:, 1],s=300, c='red', alpha=0.5)
plt.xlabel('ConsumePower'), plt.ylabel('WorkoutTimePerWeek')
plt.show()
```

-변수 중 ‘gate_30’을 ‘A시안’으로 ‘gate_40’을 ‘B시안’으로 재정의하여 AB테스트를 진행. ‘retention7’을 구매 여부로 True는 '구매한다'이고 False는 '구매하지 않는다'이다. 



```python
df_abtest.info()
```

```python
df_abtest.describe()
```

```python
#집단이 다른 두 데이터 비교를 신뢰할 수 있는지 알아보기 위해 독립표본 t-검증을 실시한다.
#변수를 분석하기 쉽게 변환한다. 
df_abtest['Creative'] = 'A'
df_abtest.loc[df_abtest['version'] == 'gate_40', 'Creative'] = 'B'
#version열의 gate_30은 시안A로, gate_40은 시안B로 간주한다.
df_abtest['Purchase'] = 0
df_abtest.loc[df_abtest['retention_7'] == True, 'Purchase'] = 1
#retention_7열의 False=0=구매안함으로, True=1=구매함으로 간주한다.
df_abtest
```

```python
#각각 시안의 구매 평균이다.
df_abtest.groupby('Creative')['Purchase'].mean()
```

 A시안(Creative A)이 더 구매가능성 높다.

```python
#A시안을 사용한 데이터
df_A = df_abtest[df_abtest['Creative'] == 'A']
print(df_A.shape)
df_A.head()
```

```python
#B시안을 사용한 데이터
df_B = df_abtest[df_abtest['Creative'] == 'B']
print(df_B.shape)
df_B.head()
```

```python
#T-test를 통해 신뢰도를 측정한다.
from scipy import stats

T_test = stats.ttest_ind(df_A['Purchase'], df_B['Purchase'])
print(T_test)
```
